```python
data_industry=data.groupby('industry').sum()
```

```python
data_industry.describe()
```

```python
data_industry_major = data_industry[data_industry['employees_on_platform']>200000000]
```

```python
plt.figure(figsize=(16,8))
ax = plt.subplot(121, aspect='equal')
data_industry_major.plot(kind='pie', y = 'employees_on_platform', ax=ax, autopct='%1.1f%%', 
 startangle=90, shadow=False, labels=data_industry_major.index, legend = False, fontsize=10)
```
